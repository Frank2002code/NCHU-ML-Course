## Exercise 1: _Load the MNIST dataset and split it into a training set and a test set (take the first 60,000 instances for training, and the remaining 10,000 for testing).(10%)

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', as_frame=False)
X_train, y_train = mnist.data[:60000], mnist.target[:60000]
X_test, y_test = mnist.data[60000:], mnist.target[60000:]
```

## Exercise 2: _Train a Random Forest classifier (estimator number=100, random state=42) on the dataset and time how long it takes, then evaluate the resulting model on the test set (use 'accuracy_score').To time how long a code runs, use '%time'.Ex: %time print('hello world')(15%)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Initialize the Random Forest classifier
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier and time the training process
%time rf_clf.fit(X_train, y_train)

# Evaluate the model
y_pred = rf_clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Test set accuracy: {accuracy}")
```

## Exercise 3: _Next, use PCA to reduce the dataset's dimensionality, with an explained variance ratio of 95%.(20%)

```python
from sklearn.decomposition import PCA

# Initialize PCA to retain 95% of the variance
pca = PCA(n_components=0.95)

# Fit PCA on the training data and transform both training and test data
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Verify the shapes of the transformed datasets
print(f"Original training set shape: {X_train.shape}")
print(f"Transformed training set shape: {X_train_pca.shape}")
```

## Exercise 4: _Train a new Random Forest classifier (estimator number=100, random state=42) on the reduced dataset and see how long it takes. Was training much faster? (The results may surprise you)(15%)

```python
rf_clf_pca = RandomForestClassifier(n_estimators=100, random_state=42)

%time rf_clf_pca.fit(X_train_pca, y_train)
```

No, it's not faster. It's slower.

## Exercise 5: _Next evaluate the classifier on the test set: how does it compare to the previous classifier? (make sure to transform the test data using 'pca.transform()')(20%)

```python
# Evaluate the model
y_pred_pca = rf_clf_pca.predict(X_test_pca)

accuracy_pca = accuracy_score(y_test, y_pred_pca)
print(f"Test set accuracy: {accuracy_pca}")
```

The accuracy is lower when not using PCA

## Exercise 6: _Try again with an `SGDClassifier` (random state=42). Print the training time and accuracy of SGD with and without PCA, respectively.(20%)

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)

%time sgd_clf.fit(X_train, y_train)

y_pred = sgd_clf.predict(X_test)
accuracy_without_pca = accuracy_score(y_test, y_pred)
print(f"Accuracy without PCA: {accuracy_without_pca}")

sgd_clf_pca = SGDClassifier(random_state=42)

%time sgd_clf_pca.fit(X_train_pca, y_train)

y_pred_pca = sgd_clf_pca.predict(X_test_pca)
accuracy_with_pca = accuracy_score(y_test, y_pred_pca)
print(f"Accuracy with PCA: {accuracy_with_pca}")
```
